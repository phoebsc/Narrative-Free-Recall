# Helper functions

```python
# for each row, record the largest number
def precise_matches_mat(corr_mat):
    temp =  np.zeros(corr_mat.shape)
    for r in range(corr_mat.shape[0]):
        temp[r,np.argmax(corr_mat, 1)[r]] = np.max(corr_mat, 1)[r]
        
    return temp

# 
def distinct_matches_mat(corr_mat):
    temp =  np.zeros(corr_mat.shape)
    std = zscore(corr_mat, 0)

    for r in range(temp.shape[0]):
        temp[r, np.argmax(std, 1)[r]] = np.max(std, 1)[r]
    return temp

def unique_match_mat(corr_mat):
    temp = find_unique_match_events(corr_mat)
    return temp

# compute unique matched events
# corrmat is a matrix with size num_chapter_events X num_recall_events
def find_unique_match_events(corrmat):
    unique_match_weighted_sim = np.copy(corrmat)

    # for each event in the chapter
    for event in range(corrmat.shape[0]):

        # loop over each event in the recall until you find the recall event
        # that is the best match for the chapter event and the chapter event is the best match for the recall event
        match = 0
        for recall_event in range(unique_match_weighted_sim.shape[1]):
            
            # get the best recall event for this chapter event
            best_recall_match = np.argmax(unique_match_weighted_sim[event,:])
            
            # get the best chapter event for the previously found best recall event
            best_book_match = np.argmax(unique_match_weighted_sim[:,best_recall_match])
            
            if best_book_match != event:
                unique_match_weighted_sim[event, best_recall_match] = 0
            else:
                match = 1
                # if the two best events match, then zero out every other recall event for the same chapter event
                for e in range(unique_match_weighted_sim.shape[1]):
                    if e != best_recall_match:
                        unique_match_weighted_sim[event, e] = 0
                break
                
        # if there was no match for this chapter event, then zero out all of the columns for it
        if match == 0:
            unique_match_weighted_sim[event, :] = 0        
    return unique_match_weighted_sim 

# aggregate the embeddings by averaging all token-level embeddings within the bounds
def mean_pool(embs, bounds):
    aggregated_embs = []
    for bound in bounds:
        aggregated_embs.append(np.mean(embs[bound[0]:bound[1],:], 0))
    return np.vstack(aggregated_embs)
   
# aggregate the embeddings by averaging all token-level embeddings within the bounds
def last_word(embs, bounds):
    aggregated_embs = []
    for bound in bounds:
        aggregated_embs.append(embs[bound[1]-1,:])  # messed up wth GPT2 (1 offset in the beginning)
    return np.vstack(aggregated_embs)    

def bert_score(embs1, bounds1, embs2, bounds2, idf1):
    sim = np.zeros((len(bounds1), len(bounds2)))
    for b1,bound1 in enumerate(bounds1):
        sent1 = embs[bound1[0]:bound1[1],:]
        idf1_sent = idf1[bound1[0]:bound1[1]]
        for b2, bound2 in enumerate(bounds2):
            sent2 = embs2[bound2[0]:bound2[1],:]
            sim[b1,b2] = compute_weighted_similarity(sent1, sent2, idf1_sent)
    return sim

def get_sentence_bounds(text_to_sample_from):
    
    import codecs
    book = ' '.join(text_to_sample_from)

    import nltk
    sentences = nltk.tokenize.sent_tokenize(book)
    print("Total Sentences are " + str(len(sentences)))

    import re
    # sentences to list of words
    sent_words = []
    total_tokens = 0
    for raw_sent in sentences:
        clean = re.sub("\n","", raw_sent)
        sent_words.append(clean.split(" "))

    words_book = np.hstack(sent_words)
    print("Total words are " + str(len(words_book)))
    
    sent_len = 0
    sent_bounds = []
    start_bound = 0
    end_bound = 0
    for sent in sent_words:
        #print(sent, len(sent))
        if sent_len+len(sent)-start_bound+1 > 10:
            end_bound = sent_len+len(sent)-1
            sent_bounds.append((start_bound, end_bound))
            #print(sent_bounds)
            sent_len = sent_len + len(sent)
            start_bound = sent_len
        else:
            sent_len = sent_len + len(sent)
            
    if len(text_to_sample_from) != sent_bounds[-1][1]-1:
        sent_bounds[-1] = (sent_bounds[-1][0], len(text_to_sample_from)-1)
    
    return sent_bounds


#### SPECIFIC TO OUR PIPELINE ######

all_words_book = np.load('/Users/mt6949/Desktop/Research/roger_ackroyd_words.npy')
chapters_info = np.load('/Users/mt6949/Desktop/Research/chapters_info.npy', allow_pickle=True).item()
# arg chapter is an int
def get_chapter_text_without_titles(words_book, chapter, chapters_dict=chapters_info['titles'], chapter_start_inds=chapters_info['start_inds'], num_to_text=chapters_info['num_to_text']):
    
    # find where the chapter starts
    chapter_title = chapters_dict[num_to_text[chapter]]
    num_words_title = len(chapter_title.split())
    chapter_start_ind = chapter_start_inds[num_to_text[chapter]]
    chapter_start_ind_without_title = chapter_start_ind + 1 + num_words_title # add 1 for the number of the chapter

    # find where the chapter ends
    if chapter == 27: # if last chapter, then end index is the end of the book
        chapter_end_ind = len(words_book)
    else:
        chapter_end_ind = chapter_start_inds[num_to_text[chapter+1]]

    chapter_text_without_title = words_book[chapter_start_ind_without_title:chapter_end_ind]
    return chapter_text_without_title 

# arg chapters is a list or an array of ints
def text_from_multiple_chapters(words_book, chapters):

    all_text_without_titles = []
    for chapter in chapters:
        chapter_text_without_title = get_chapter_text_without_titles(words_book, chapter)
        all_text_without_titles.append(chapter_text_without_title)

    return np.hstack(all_text_without_titles)

########################################
```

# Compute stimulus-recall pairwise correlations

```python
# assumes you already have some semantic representations of the stimulus and the recall

stimulus_embs = np.load('/Users/mt6949/Desktop/TopicModels/chapter{}_model_50wsize.npy'.format(chapter))  
stimulus_sentence_bounds = get_sentence_bounds(stimulus_text)
aggregate_stimulus_embs = last_word(stimulus_embs, stimulus_sentence_bounds)

recall_corrs = OrderedDict()
for sub in range(1,30):

    recall_embs = np.load('/Users/mt6949/Desktop/TopicModels/gpt_{}_generations_{}_model_400topics_50booksize_50recallsize.npy'.format(model_type,sub))
    sentence_bounds = get_sentence_bounds(gpt_generations[sub].split(' '))
    aggregate_recall_embs = last_word(recall_embs, sentence_bounds)

    recall_corrs[sub] = 1-cdist(aggregate_stimulus_embs,aggregate_recall_embs,'correlation')

    print('done with sub {}'.format(sub))
    
    plt.matshow(recall_corrs[sub])
    plt.colorbar()
    plt.clim([-1,1])
    plt.show()  
```

# Apply metrics on top of pairwise correlation matrix

```python
gpt_generations_precise_matches = np.zeros((len(recall_corrs),aggregate_stimulus_embs.shape[0]))

for sub in recall_corrs.keys():
    gpt_generations_precise_matches[sub-1,:] = np.sum(unique_match_mat(recall_corrs[sub]),1)
```
